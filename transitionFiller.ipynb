```python
binary_values = [
    "0000",
    "0000",
    "0000",
    "0001",
    "0010",
    "0100", 
    "0110",
    "0111",
    "1101",
    "1111",
    "1111",
    "1111"
]

# Each value has 16 bits; bit 15 is MSB (leftmost)
bit_transitions = [[] for _ in range(4)]

# Fill transitions for each bit
for value in binary_values:
    for i in range(4):
        bit_transitions[i].append(int(value[i]))  # MSB at index 0

interval = 80.0
#final_interval = 80.0

for bit_index in range(4):
    bit_number = 3 - bit_index  # Reverse index to match inR1[15] ... inR1[0]
    transitions = bit_transitions[bit_index]
    print(f'TRANSITION_LIST("opcode[{bit_number}]") {{')
    print('    NODE')
    print('    {')
    print('        REPEAT = 1;')

    current_level = transitions[0]
    duration = interval

    for level in transitions[1:]:
        if level == current_level:
            duration += interval
        else:
            print(f'        LEVEL {current_level} FOR {duration};')
            current_level = level
            duration = interval

    print(f'        LEVEL {current_level} FOR {duration};')
    #print(f'        LEVEL 0 FOR {final_interval};')
    print('    }')
    print('}\n')
```
